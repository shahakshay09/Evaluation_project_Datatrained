### Removing Skewness using yeo-johnson method

```python
skewd = X[['DistanceFromHome', 'MonthlyIncome', 'TotalWorkingYears', 'YearsAtCompany','PercentSalaryHike','YearsWithCurrManager','YearsInCurrentRole']]
```

```python
from sklearn.preprocessing import PowerTransformer
power = PowerTransformer(method='yeo-johnson')
```

```python
transfrm =  power.fit_transform(skewd)
```

```python
skewd_data = pd.DataFrame(transfrm, columns=skewd.columns) #to convert numpy array back into dataframe
```

```python
skewd_data.skew()
```

```python
skewd_data.index = X.index
```

```python
X[['DistanceFromHome', 'MonthlyIncome', 'TotalWorkingYears', 'YearsAtCompany','PercentSalaryHike','YearsWithCurrManager','YearsInCurrentRole']] = skewd_data[['DistanceFromHome', 'MonthlyIncome', 'TotalWorkingYears', 'YearsAtCompany','PercentSalaryHike','YearsWithCurrManager','YearsInCurrentRole']]
```

# Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
maxAcc = 0
maxRS=0
for i in range(1,100):
    x_train,x_test,y_train,y_test = train_test_split(scaled_x_best,y,test_size = .31, random_state = i,stratify = y)
    modRF =  LogisticRegression()
    modRF.fit(x_train,y_train)
    pred = modRF.predict(x_test)
    acc  = accuracy_score(y_test,pred)
    if acc>maxAcc:
        maxAcc=acc
        maxRS=i
print(f"Best Accuracy is: {maxAcc} on random_state: {maxRS}")
```
